```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
user = pd.read_csv('../data/dup/user.csv')
user.head()
```

```python
user['residenceProv'] = user['residence'].apply(lambda x:int(x/100))
user.head()
```

```python
plt.figure(figsize=(18,6))
sns.distplot(user['age'])
plt.show()
```

```python
plt.figure(figsize=(18,6))
t1 = user[user['haveBaby']==1]['age']
t2 = user[user['haveBaby']==2]['age']
t3 = user[user['haveBaby']==3]['age']
t4 = user[user['haveBaby']==4]['age']
t5 = user[user['haveBaby']==5]['age']
t6 = user[user['haveBaby']==6]['age']
sns.kdeplot(t1,label='1')
sns.kdeplot(t2,label='2')
sns.kdeplot(t3,label='3')
sns.kdeplot(t4,label='4')
sns.kdeplot(t5,label='5')
sns.kdeplot(t6,label='6')
plt.show()
```

```python
plt.figure(figsize=(18,6))
t1 = user[user['marriageStatus']==1]['age']
t2 = user[user['marriageStatus']==2]['age']
t3 = user[user['marriageStatus']==3]['age']

sns.kdeplot(t1,label='1')
sns.kdeplot(t2,label='2')
sns.kdeplot(t3,label='3')

plt.show()
```
