```python
user = pd.read_csv('../data/dup/user.csv')
user.head()
```

```python
d = d.merge(user,on='userID',how='left')
d.head()
```

```python
data = d[d['label']==0]
sns.kdeplot(data['age'],label='label 0')
data = d[d['label']==1]
sns.kdeplot(data['age'],label='label 1')
```

```python
dt = d['gender']
dt.head()
```

```python
t = pd.get_dummies(dt,prefix='gender')
t.head()
```

```python
cols = ['clickTime','connectionType','telecomsOperator','education','marriageStatus','haveBaby']
for col in cols:
    dt = d[col]
    t = pd.get_dummies(dt,prefix=col)
    d.drop(col,axis=1,inplace=True)
    d = d.join(t)
d.head()
```
